```python
# for each patinet this function genrates a list containing the most compatible donor i.e. highest time till rejection
def best_donor(patient):
    data = df.to_numpy()
    patient =int(patient)
    donor_list =[]
    for idx, x in np.ndenumerate(data):
        if idx[0] == patient -1:
            donor_list.append([x, idx[1]+1, patient])
    donor_list = sorted(donor_list, reverse=True)
    donor_list = [ elem for elem in donor_list if elem[0] != 0]
    donor_list[0].reverse()
    return donor_list[0]

# for each donor this function genrates a list containing the most compatible patient i.e. highest time till rejection 
def best_patient(donor):
    data = df.T.to_numpy()
    donor =int(donor)
    p_list =[]
    for idx, x in np.ndenumerate(data):
        if idx[0] == donor -1:
            p_list.append([x, idx[1]+1, donor])
    p_list = sorted(p_list, reverse=True)
    p_list = [ elem for elem in p_list if elem[0] != 0]
    p_list[0].reverse()
    return p_list[0]

print(best_donor(1))
print(best_patient(1))
```

```python
# This function generates all possible chains containing the patient-donor pairs that have not been dealth with via cycles 
# i.e pairs that are free. Takes new_list as an input.
# This function takes 1-2 mins to run the first time since it has (300-40) 260 pairs to process but gets faster 
# after every successive iteration.
# [In this case when dealing with chains we have to re-calculate all the available chains since after we process each chain
# not all members in the chain are accounted for and this function has to be called everytime the new_list gets updated.]
# This function also sorts the chains according the descending order of their length which makes implementing the 
# first_tie breaking rule of longest chain first easier later on.
def mod_chains(n_list):
    chain_list =[]
    for i in new_list:
        b_list =[i]
        while any(b_list.count(element) > 1 for element in b_list) == False :
            #print('yes')
            b_list.append(best_patient(b_list[-1])[1])
            if b_list[-1] not in new_list:
                b_list.pop()
                break
        #print(b_list)
        chain_list.append(b_list)
        for chain in chain_list:
            #removes chains with only one element
            if len(chain) == 1 or len(chain) == 2 :
                chain_list.remove(chain)
        chain_list = sorted(chain_list, key =len)
        chain_list.reverse()
    return chain_list
```

```python
# when we process a chain the except for the first and last member of the chain every other pair is completed
# i.e the patient in that pair has received a kidney and the corresponding donor has donated their kidney to another
# patient. for the first member of a chain we have an un-assigned patient and for the last member of the chain we have
# a free donor. The unassigned donor get's added to the unassigned_patient list / waiting list and the free donor donates
# gets added to the free_donor list. We will process these two lists after we have dealt with all possible chains. 
unassigned_patient = []
free_donor = []

# This function processes a chain i.e adds the first memeber to unassigned list, the last member to the free donor list and
# updates the list of pairs allready processed so that we can create new chains with only the pairs that have not been 
# assigned a kidney and does not have a free donor.
# f_list => finished list stands for all the patient-donor pairs where both the patient and donor have been assigned 
# so no one is free and new_list all the patient-donor pairs where both the patient and donor are free

def process_chain(chain):
    unassigned_patient.append(chain[0])
    free_donor.append(chain[-1])
    for x in chain[1:-1]:
        f_list.append(x)
    for x in chain:
        new_list.remove(x)
    return new_list
```

```python
#calculates the maximal distance of a chain as this serves as a tie-breaking rule
def calculate_max_chain(chain):
    chain = list(chain)
    max_val = []
    i = 0
    for i in range(0,len(chain) -1):
        max_val.append(best_patient(chain[i])[2])
    return sum(max_val)

print(calculate_max_chain([299, 180]))
```

```python
# This is where we implement the tie-breaking rules
# shortest maximal chain
# chains starting with donor who has blood type "O" 
def chain_selector(n_list):
    dummy_list = []
    dummy_list2 = []
    chain_s = mod_chains(new_list)
    for chain in chain_s:
        if len(chain) == len(chain_s[0]):
            dummy_list.append(chain)
    if len(dummy_list) == 1:
        return dummy_list[0]
    elif len(dummy_list) > 1:
            i = 0
            for i in range(0, len(dummy_list)):
                dummy_list2.append([calculate_max_chain(dummy_list[i]),dummy_list[i]])
            dummy_list2 = sorted(dummy_list2, reverse = True)
            if dummy_list2[0][0] > dummy_list2[1][0]:
                return dummy_list2[0][1]
            elif dummy_list2[0][0] == dummy_list2[1][0]:
                if blood_type(best_donor(dummy_list2[0][1][0])[1])=='O':
                    return dummy_list2[0][1]
                elif blood_type(best_donor(dummy_list2[1][1][0])[1])=='O':
                    return dummy_list2[1][1]
                else:
                    return dummy_list2[0][1]
            else:
                return dummy_list2[0][1]
    elif len(dummy_list) == 0:
        return "No chain"

```
